```python
import os
import json

import numpy as np
import pandas as pd

from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler

import matplotlib.pyplot as plt
# plt.rcParams['figure.figsize'] = [30, 30]
```

```python

root_dir = "."

input_dir = os.path.join(root_dir, "input")
output_dir = os.path.join(root_dir, "output")
```

```python

scoring_json = os.path.join(input_dir, "I-P_decade_scoring_ChatGPT_extrapolated.json")

with open(scoring_json, "r") as in_json:
    scoring_dict = json.load(in_json)
```

```python
scoring_criteria = [
    'Historical Claims', 
    'Security and Peace Efforts', 
    'Legal and International Standing', 
    'Humanitarian Impact', 
    'Diplomatic Achievements', 
    'Narrative and Media Influence', 
    'Strategic Positioning', 
    'International Alliances', 
    'Economic Impact', 
    'Social Cohesion', 
    'Innovation and Resilience', 
    'Environmental Sustainability', 
    'Cultural Influence', 
    'Legal System Integrity', 
    'International Sympathy', 
    'Capacity for Reform and Innovation', 
    'Resolution Feasibility', 
    'International Law Compliance', 
    'Human Rights Record', 
    'Efforts Towards Peace', 
    'Impact on Civilians', 
    'International Support and Perception', 
    'Cultural and Religious Respect', 
    'Economic Stability and Development', 
    'Internal Unity and Political Stability', 
    'Transparency and Accountability'
]
```

```python

dataframe_dict_i = {}
dataframe_dict_p = {}

for criteria in scoring_criteria:

    dataframe_dict_i[criteria] = []
    dataframe_dict_p[criteria] = []
```

```python

decades = [decade for decade in scoring_dict.keys() if decade not in [
    # "1880s", "1890s" # , "1900s" # , "1910s", "1920s", "1930s"
]]

for decade in decades:

    for criteria in scoring_criteria:
        
        dataframe_dict_i[criteria].append(scoring_dict[decade]["israeli"][criteria])
        dataframe_dict_p[criteria].append(scoring_dict[decade]["palestinian"][criteria])
```

```python

i_scoring_df = pd.DataFrame(dataframe_dict_i)

# i_scoring_df[scoring_criteria] = StandardScaler().fit_transform(i_scoring_df)
i_scoring_df[scoring_criteria] = MinMaxScaler(
    feature_range=(1E-1,1-(1E-1))
).fit_transform(i_scoring_df)

i_scoring_df.head()
```

```python

p_scoring_df = pd.DataFrame(dataframe_dict_p) 

# p_scoring_df[scoring_criteria] = StandardScaler().fit_transform(p_scoring_df)
p_scoring_df[scoring_criteria] = MinMaxScaler(
    feature_range=(1E-1,1-(1E-1))
).fit_transform(p_scoring_df)

p_scoring_df.head()
```

```python

swing_comparison_df = pd.DataFrame(
    (i_scoring_df.values / p_scoring_df.values) - 1, columns = scoring_criteria)

swing_comparison_df["decades"] = decades
swing_comparison_df = swing_comparison_df.set_index("decades")

swing_comparison_df.head() 
```

```python

def lower_5(x):
    return x.quantile(0.05)
def lower_q(x):
    return x.quantile(0.25)
def upper_q(x):
    return x.quantile(0.75)
def upper_5(x):
    return x.quantile(0.95)
    
stats_df = swing_comparison_df.agg(
    ['mean', 'min', 'max', 'quantile', 'var', 'median', 'std', lower_5, lower_q, upper_q, upper_5],
    axis=1
)

stats_df.head()
```

```python

plt.figure(figsize=(12, 8))
plt.plot(stats_df.index, stats_df['lower_5'], label='Pro Pal fifth percentile', linestyle='--')
plt.plot(stats_df.index, stats_df['mean'] - stats_df['std'], label='Pro Pal std', linestyle=':')
plt.plot(stats_df.index, stats_df['mean'], label='Avg swing', linestyle='-')
plt.plot(stats_df.index, stats_df['mean'] + stats_df['std'], label='Pro Isr fifth percentile', linestyle=':')
plt.plot(stats_df.index, stats_df['upper_5'], label='Pro Isr max', linestyle='-.')
plt.plot(stats_df.index, np.zeros(len(stats_df)), label='Draw', linestyle='-.')
plt.fill_between(stats_df.index, stats_df['lower_5'], stats_df['upper_5'], alpha=0.2)
plt.title('Israel-Palestine debate swing')
plt.xlabel('Decades')
plt.ylabel('Debate score swing')
plt.legend()
plt.grid(True)

graphic_png = os.path.join(output_dir, "I-P_debate_swing.png")
plt.savefig(graphic_png)

plt.show()
```
